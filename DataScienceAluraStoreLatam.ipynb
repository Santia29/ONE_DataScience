### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



###En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
ingreso_total_tienda_1 = tienda['Precio'].sum()
ingreso_total_tienda_2 = tienda2['Precio'].sum()
ingreso_total_tienda_3 = tienda3['Precio'].sum()
ingreso_total_tienda_4 = tienda4['Precio'].sum()

stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_total_tienda_1, ingreso_total_tienda_2, ingreso_total_tienda_3, ingreso_total_tienda_4]

plt.bar(stores, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.ylabel('Ingreso Total ($)')
plt.title('Ingreso Total por Tienda')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

## 2. Ventas por categoría

###En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
ventas_por_categoria_tienda_1 = tienda['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda_2 = tienda2['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda_3 = tienda3['Categoría del Producto'].value_counts()
ventas_por_categoria_tienda_4 = tienda4['Categoría del Producto'].value_counts()

ventas_top_tienda_1 = ventas_por_categoria_tienda_1.head(5)
ventas_top_tienda_2 = ventas_por_categoria_tienda_2.head(5)
ventas_top_tienda_3 = ventas_por_categoria_tienda_3.head(5)
ventas_top_tienda_4 = ventas_por_categoria_tienda_4.head(5)

ventas_top_tienda_1.plot(kind='bar', color='skyblue')
plt.title('Top 5 Categorías de Productos - Tienda 1')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.show()

ventas_top_tienda_2.plot(kind='bar', color='lightcoral')
plt.title('Top 5 Categorías de Productos - Tienda 2')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.show()


ventas_top_tienda_3.plot(kind='bar', color='lightgreen')
plt.title('Top 5 Categorías de Productos - Tienda 3')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.show()

ventas_top_tienda_4.plot(kind='bar', color='gold')
plt.title('Top 5 Categorías de Productos - Tienda 4')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.show()
```

# 3. Calificación promedio de la tienda


#####En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python

calificacion_promedio_tienda_1 = tienda['Calificación'].mean()
calificacion_promedio_tienda_2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda_3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda_4 = tienda4['Calificación'].mean()
```

```python
calificaciones_promedio = [
    calificacion_promedio_tienda_1,
    calificacion_promedio_tienda_2,
    calificacion_promedio_tienda_3,
    calificacion_promedio_tienda_4
]
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(stores, calificaciones_promedio, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.show()
```

# 4. Productos más y menos vendidos

####En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python

productos_vendidos_tienda_1 = tienda['Producto'].value_counts()
productos_vendidos_tienda_2 = tienda2['Producto'].value_counts()
productos_vendidos_tienda_3 = tienda3['Producto'].value_counts()
productos_vendidos_tienda_4 = tienda4['Producto'].value_counts()

mas_vendidos_tienda_1 = productos_vendidos_tienda_1.head()
mas_vendidos_tienda_2 = productos_vendidos_tienda_2.head()
mas_vendidos_tienda_3 = productos_vendidos_tienda_3.head()
mas_vendidos_tienda_4 = productos_vendidos_tienda_4.head()

productos_menos_vendidos_tienda_1 = productos_vendidos_tienda_1.tail()
productos_menos_vendidos_tienda_2 = productos_vendidos_tienda_2.tail()
productos_menos_vendidos_tienda_3 = productos_vendidos_tienda_3.tail()
productos_menos_vendidos_tienda_4 = productos_vendidos_tienda_4.tail()
```

```python

mas_vendidos_tienda_1.plot(kind='barh', color='skyblue')
plt.title('Top 5 Productos Más Vendidos - Tienda 1')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()

productos_menos_vendidos_tienda_1.plot(kind='barh', color='lightcoral')
plt.title('Top 5 Productos Menos Vendidos - Tienda 1')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()

mas_vendidos_tienda_2.plot(kind='barh', color='skyblue')
plt.title('Top 5 Productos Más Vendidos - Tienda 2')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()

productos_menos_vendidos_tienda_2.plot(kind='barh', color='lightcoral')
plt.title('Top 5 Productos Menos Vendidos - Tienda 2')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()

mas_vendidos_tienda_3.plot(kind='barh', color='skyblue')
plt.title('Top 5 Productos Más Vendidos - Tienda 3')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.tight_layout()
plt.show()

productos_menos_vendidos_tienda_3.plot(kind='barh', color='lightcoral')
plt.title('Top 5 Productos Menos Vendidos - Tienda 3')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()

mas_vendidos_tienda_4.plot(kind='barh', color='skyblue')
plt.title('Top 5 Productos Más Vendidos - Tienda 4')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()

productos_menos_vendidos_tienda_4.plot(kind='barh', color='lightcoral')
plt.title('Top 5 Productos Menos Vendidos - Tienda 4')
plt.xlabel('Cantidad de Ventas')
plt.ylabel('Producto')
plt.show()
```

# 5. Envío promedio por tienda

####En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python

envio_promedio_tienda_1 = tienda['Costo de envío'].mean().round(1)
envio_promedio_tienda_2 = tienda2['Costo de envío'].mean().round(1)
envio_promedio_tienda_3 = tienda3['Costo de envío'].mean().round(1)
envio_promedio_tienda_4 = tienda4['Costo de envío'].mean().round(1)
```

```python
envios_promedio = [
    envio_promedio_tienda_1,
    envio_promedio_tienda_2,
    envio_promedio_tienda_3,
    envio_promedio_tienda_4
]
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.bar(stores, envios_promedio, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.title('Costo de Envío Promedio por Tienda')
plt.ylabel('Costo de Envío Promedio ($)')
plt.show()
```
